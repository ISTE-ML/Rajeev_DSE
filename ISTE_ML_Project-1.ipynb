### Using inbuilt Pandas function(**read_csv**)

```python
cars = pd.read_csv('/content/mtcars.csv')
cars.head()
```

```python
length = cars.shape[0]
# In this file, length = 32
```

### **DESC**

```python
cars.describe()
```

### **Method 2**

```python
print(cars.count())
print(cars.mean())
print(cars.std())
print(cars.min())
print(cars.quantile())
print(cars.max())
```

### **Method 3**

```python
for i in range(1, cars.shape[1]):
  s = 0.0
  x = cars.iloc[:, i].tolist()
  x_sum = sum(x)
  x_mean = x_sum / length
  for j in range(0, cars.shape[0]):
    s += ((x[j] - x_mean)**2)
  print(f"""
  {cars.columns.values[i]} :
  Count = {length}
  Mean = {x_mean}
  Minimum = {min(x)}
  Maximum = {max(x)}
  Standard Deviation = {(s / (length-1)) ** (0.5)}
  """)
```
